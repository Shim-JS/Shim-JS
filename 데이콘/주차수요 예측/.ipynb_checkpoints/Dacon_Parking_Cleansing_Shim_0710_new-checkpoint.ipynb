```python
# 데이터 로드
train = pd.read_csv('data/train0702.csv')
test = pd.read_csv('data/test0702.csv')
```

#### 지하철, 버스 Rename

```python
# 지하철, 버스 rename
train.rename(columns={"도보 10분거리 내 지하철역 수(환승노선 수 반영)" : "지하철", "도보 10분거리 내 버스정류장 수" : "버스"}, inplace=True)
test.rename(columns={"도보 10분거리 내 지하철역 수(환승노선 수 반영)" : "지하철", "도보 10분거리 내 버스정류장 수" : "버스"}, inplace=True)
```

- 버스정류장 경상북도, 비슷한 임대료의 집들을 확인하여 6의 값으로 대치 예정

- 지하철 충청남도, 경상남도는 지하철이 존재하지 않는다. => 따라서 0으로 대치하는 것이 좋아 보임.

- 지하철 대전광역시는 0으로 대치해야하나? 뭐로 대치해야하나 0으로도 대치 1로도 대치

```python
# train 데이터셋 지하철, 버스 데이터 전처리
train["지하철"] = train["지하철"].fillna(0).astype("object")
train["버스"] = train["버스"].fillna(6).astype("object")
```

```python
# 추가적인 결측치 개수 파악
pd.DataFrame(train.isna().sum()).T
```

### 데이터 합치기

```python
# 데이터 로드
age_gender = pd.read_csv('data/age_gender_info.csv')
train = pd.read_csv('data/train0710.csv', encoding = 'ansi')
test = pd.read_csv('data/test0710.csv', encoding = 'ansi')
```

```python
train_m = pd.merge(train,age_gender,how = 'left', on = '지역')
test_m = pd.merge(test,age_gender,how = 'left', on = '지역')
```

```python
train_m.info()
test_m.info()
```

```python
train_m.to_csv("trainm0710.csv", encoding='ansi', index = None)
test_m.to_csv("testm0710.csv", encoding='ansi', index = None)
```
