```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from matplotlib import font_manager, rc
font_path = "C:/Windows/Fonts/NGULIM.TTF"
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)

import warnings
warnings.filterwarnings("ignore")
```

```python
age_gender = pd.read_csv('data/age_gender_info.csv')
train = pd.read_csv('data/train1.csv', encoding = "ansi")
test = pd.read_csv('data/test3.csv', encoding = "ansi")
```

```python
#지하철역수 결측치 0으로 대치
train["지하철역수"] = train["지하철역수"].fillna(0)
test["지하철역수"] = test["지하철역수"].fillna(0)
```

```python
#자격유형 결측치 확인
test[test["자격유형"].isna()]
```

```python
test[test["단지코드"] == "C2253"]
```

```python
# 자격유형 결측치 처리
test.iloc[196] = test.iloc[196].fillna("A")
test.iloc[258] = test.iloc[258].fillna("C")
```

```python
#지하철역수 , 버스정류장수 float ==> object 
train["지하철역수"] = train["지하철역수"].astype(object)
train["버스정류장수"] = train["버스정류장수"].astype(object)
test["지하철역수"]= test["지하철역수"].astype(object)
test["버스정류장수"] = test["버스정류장수"].astype(object)
```
