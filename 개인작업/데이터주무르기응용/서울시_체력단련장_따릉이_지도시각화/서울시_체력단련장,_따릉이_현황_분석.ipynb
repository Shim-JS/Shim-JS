## 1) 데이터 로드


```python
import pandas as pd

df = pd.read_csv("/content/drive/Shareddrives/금융공모전/심종수/체력단련장업.csv",encoding = "cp949")
df
```

```python
df.info()
```

```python
# 데이터 프레임 재설정
df_1 = df.loc[:,["번호","사업장명","영업상태명","소재지전체주소","좌표정보(x)","좌표정보(y)"]]
df_1 = df_1.dropna()
df_1
```

```python
## 영업중인 체력단련장 추출
df_2 = df_1.loc[df_1["영업상태명"] == "영업/정상"]
df_2
```

```python
## 소재지 전체 주소에서 서울특별시를 설정
df_3 = df_2.loc[df_2["소재지전체주소"].str.contains('서울',na=False)]
df_3
```

```python
import googlemaps

gkeys = "**************************"
gmaps = googlemaps.Client(key=gkeys)
```

```python
# 구글맵 활용
gmaps.geocode('코엑스짐 중곡점', language='ko')
```

```python
# 서울 체력단련장 이름 변경
juso_name = []

for name in df_3['소재지전체주소']:
    juso_name.append('대한민국 ' + str(name))

juso_name
```

```python
station_addreess = []
station_lat = []
station_lng = []
# 경찰서의 이름으로 주소 가져오기
for name in juso_name:
    tmp = gmaps.geocode(name, language='ko')
    station_addreess.append(tmp[0].get("formatted_address"))
    
    tmp_loc = tmp[0].get("geometry")

    station_lat.append(tmp_loc['location']['lat'])
    station_lng.append(tmp_loc['location']['lng'])
    
    print(name + '-->' + tmp[0].get("formatted_address"))
```

```python
df_3['lat'] = station_lat
df_3['lng'] = station_lng
```

```python
df_3
```

```python
df_3.reset_index()
df_3.drop("index",axis = 0)
```

```python
import folium
```

```python
# 체력단련장,따릉이의 위치 확인
map = folium.Map(location=[37.5502, 126.982], zoom_start=11)

for n in df_3.index:
    folium.Marker([df_3['lat'][n], 
                   df_3['lng'][n]],
                  icon=folium.Icon('red', icon='star')).add_to(map)

for n in fd.index:
    folium.Marker([fd['위도'][n], 
                   fd['경도'][n]]).add_to(map)
    
```

```python
map.save("/content/drive/Shareddrives/금융공모전/심종수/체력단련장업_따릉이.html")
```
