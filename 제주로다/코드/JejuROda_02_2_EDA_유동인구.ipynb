```python
# 데이터 불러오기
korea = pd.read_csv("KOREAN.CSV", encoding = "cp949")
korea.tail(3)
```

```python
# 거주인구 성별 확인
korea.groupby('sex')['resd_pop_cnt'].mean().plot(kind = "bar", figsize = (5,3), fontsize = 15, width=0.7, color="lightslategrey")

plt.xlabel('', fontsize = 13)
plt.xticks(rotation = 0)
plt.ylabel('resd_pop_cnt', fontsize = 13)
plt.legend(["거주인구"])
plt.title("성별 (평균)", fontsize = 20)
plt.show()
```

```python
# 방문인구 성별 확인
korea.groupby('sex')['visit_pop_cnt'].mean().plot(kind = "bar", figsize = (5,3), fontsize = 15, width=0.7, color="lightslategrey")

plt.xlabel('', fontsize = 13)
plt.xticks(rotation = 0)
plt.ylabel('visit_pop_cnt', fontsize = 13)
plt.legend(["방문인구"])
plt.title("성별 (평균)", fontsize = 20)
plt.show()
```

```python
# 국적별 방문인구 확인
LONG_TERM_FRGN.groupby('nationality')['visit_pop_cnt'].mean().sort_values("index",ascending=False).plot(kind = "bar", figsize = (5,3), fontsize = 15, width=0.7, color="lightslategrey")

plt.xlabel('', fontsize = 13)
plt.xticks(fontsize = 10, rotation = 90)
plt.ylabel('visit_pop_cnt', fontsize = 13)
plt.legend(["방문인구"],loc="upper right")
plt.title("국적 (평균)", fontsize = 20)
plt.show()
```

```python
# 년-월만 추출
em_cnt_M = LONG_TERM_FRGN["base_date"]

c3_list = []
for i in em_cnt_M:
    i = str(i)  # 숫자 -> 문자로 변환
    i = i[0:7]  # month 부분만 추출
    c3_list.append(i)  # list 로 저장 
#print(c2_list)

LONG_TERM_FRGN.insert(1,"base_date_ym",c3_list)
```

```python
LONG_TERM_FRGN["people"] = LONG_TERM_FRGN["resd_pop_cnt"] + LONG_TERM_FRGN["work_pop_cnt"] + LONG_TERM_FRGN["visit_pop_cnt"]
LONG_TERM_FRGN.head(3)
```

#### 단기체류 외국인 데이터

```python
SHORT_TERM_FRGN = pd.read_csv("SHORT_TERM_FRGN.CSV", encoding = "cp949")
SHORT_TERM_FRGN.tail(3)
```

```python
# 국적별 방문인구 확인
SHORT_TERM_FRGN.groupby('nationality')['visit_pop_cnt'].mean().sort_values("index",ascending=False).plot(kind = "bar", figsize = (5,3), fontsize = 15, width=0.7, color="lightslategrey")

plt.xlabel('', fontsize = 13)
plt.xticks(fontsize = 10, rotation = 90)
plt.ylabel('visit_pop_cnt', fontsize = 13)
plt.legend(["방문인구"],loc="upper right")
plt.title("국적 (평균)", fontsize = 20)
plt.show()
```

```python
# 시간대별 방문인구 확인
SHORT_TERM_FRGN.groupby('time')['visit_pop_cnt'].sum().plot(kind = "bar", figsize = (5,3), fontsize = 15, width=0.7, color="lightslategrey")

plt.xlabel('', fontsize = 13)
plt.xticks(fontsize = 11, rotation = 0)
plt.ylabel('work_pop_cnt', fontsize = 13)
plt.legend(["방문인구"],loc="upper right")
plt.title("시간대", fontsize = 20)
plt.show()
```

```python
# 시간대별로 중복이 있으므로 일별 평균 후 -> 월별 합

LONG_TERM_FRGN_ = LONG_TERM_FRGN["mean_visit"] = LONG_TERM_FRGN.groupby(["base_date"])["visit_pop_cnt"].mean()

SHORT_TERM_FRGN_ = SHORT_TERM_FRGN["mean_visit"] = SHORT_TERM_FRGN.groupby(["base_date"])["visit_pop_cnt"].mean()
LONG_TERM_FRGN_ = LONG_TERM_FRGN_.reset_index()
SHORT_TERM_FRGN_ = SHORT_TERM_FRGN_.reset_index()
```

```python
# 년-월만 추출
em_cnt_M = LONG_TERM_FRGN_["base_date"]

c3_list = []
for i in em_cnt_M:
    i = str(i)  # 숫자 -> 문자로 변환
    i = i[0:7]  # month 부분만 추출
    c3_list.append(i)  # list 로 저장 
#print(c2_list)

LONG_TERM_FRGN_.insert(1,"base_date_ym",c3_list)

em_cnt_M = SHORT_TERM_FRGN_["base_date"]

c3_list = []
for i in em_cnt_M:
    i = str(i)  # 숫자 -> 문자로 변환
    i = i[0:7]  # month 부분만 추출
    c3_list.append(i)  # list 로 저장 
#print(c2_list)

SHORT_TERM_FRGN_.insert(1,"base_date_ym",c3_list)
```

```python
# 단기&장기체류 외국인 방문인구 그래프 확인
plt.figure(figsize=(15,4))  # size 지정

plt.plot(LONG_TERM_FRGN_["base_date_ym"].unique(), LONG_TERM_FRGN_.groupby('base_date_ym')['visit_pop_cnt'].sum(), color = 'lightslategrey')
plt.plot(LONG_TERM_FRGN_["base_date_ym"].unique(), SHORT_TERM_FRGN_.groupby('base_date_ym')['visit_pop_cnt'].sum(), color = 'green')

plt.xticks(fontsize = 11, rotation = 45)
plt.ylabel("", fontsize = 13)
plt.legend(["장기체류 외국인 방문인구","단기체류 외국인 방문인구"], loc = "upper right")
plt.title("일평균 유동인구 (합계)", fontsize = 18)
plt.show()
```
