```python
df_sns = pd.read_csv("test/sns통합.csv", encoding = "cp949")
df_sns.head()
```

```python
sns_temp = df_sns.loc[:,["날짜","내용"]]
sns_temp.head()
```

```python
sns_g = sns_temp.groupby("날짜").count()
sns_g = sns_g.reset_index()
sns_g = sns_g.rename(columns={"내용" : "SNS"})
sns_g
```

```python
def count(df1,i):
    display(df1.Title)
    
    sns_g = sns_temp.groupby("날짜").count()
    sns_g = sns_g.reset_index()
    sns_g = sns_g.rename(columns={"내용" : "SNS"})
    
    k_w = input('문자열을 입력하세요: ')
    k_c = k_w.split(",")
    sns_k = sns_temp.loc[sns_temp['내용'].str.contains('|'.join(k_c))]
    sns_k = sns_k.groupby("날짜").count()
    sns_k = sns_k.reset_index()
    sns_k = sns_k.rename(columns={"내용" : "SNS_KeyWord"})
    #display(sns_k.head(20))
    
    sns_t = sns_g.copy()
    #sns_t["SNS_KeyWord"] = 0
    sns_t = pd.merge(sns_t,sns_k,how = "left", on = "날짜") 
    sns_t = sns_t.fillna(0)
    
    df1.Date = pd.to_datetime(df1.Date)
    sns_t.날짜 = pd.to_datetime(sns_t.날짜)
    
    sns_t = sns_t.loc[sns_t["날짜"].values >= df1.기간A시작]
    sns_t = sns_t.loc[sns_t["날짜"].values <=df1.기간C종료]
    
    sns_t["분류"] = df1.Class
    sns_t["내용"] = df1.Title
    
    name = df1.Class.split("/")[0] 
    sns_t.to_csv("{}_{}.csv".format(name,i),encoding = "cp949",index = False)
    
    return sns_t
```
