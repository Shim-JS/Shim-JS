```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('sns데이터/2021(0831).xlsx')
display(df.info())
display(df.head())
```

```python
df = df.dropna(subset = ['字段1'])
```

```python
df = df.reset_index()
df
```

```python
df = df.fillna("Mola")
df
```

```python
for i in range(len(df)):
    df.loc[i,"내용"] = df.iloc[i,2] +' ' +  df.iloc[i,3] + ' ' + df.iloc[i,4]
```

```python
df = df.rename(columns={'字段1':'날짜', '标题':"아이디"})
```

```python
df = df.loc[:,["아이디","날짜","내용"]]
df
```

```python
df.내용 = df.내용.astype(str)
df.날짜 = df.날짜.astype(str)
df_1 = df
df_1['내용'] = df_1['내용'].str.replace('[^ㄱ-ㅎㅏ-ㅣ가-힣]',' ',regex=True)
df_1['내용'] = df_1['내용'].replace({'': np.nan})
df_1['내용'] = df_1['내용'].replace(r'^\s*$', None, regex=True)
#clean_Data.dropna(how='any', inplace=True)
#df_1 = df_1.reset_index (drop = True)
df_1
```

```python
#df_1['날짜'] = df_1.apply(lambda x : x['날짜'].replace("년 ",'-'),axis = 1)
df_1['year'] = '2021'
df_1['날짜'] = df_1.apply(lambda x : x['날짜'].replace("월 ",'-'),axis = 1)
df_1['날짜'] = df_1.apply(lambda x : x['날짜'].replace("일",''),axis = 1)
df_1
```

```python
df_1["날짜"].unique()
```

```python
# df_1['year'] = df_1["날짜"].apply(lambda x: x.split('-')[0])
# display(df_1)
df_1['month'] = df_1["날짜"].apply(lambda x: "{:0>2s}".format(x.split('-')[0]))
display(df_1)
df_1['day'] = df_1["날짜"].apply(lambda x: "{:0>2s}".format(x.split('-')[1]))

df_1["날짜"] = df_1['year']+'-'+df_1['month'] + '-' + df_1['day']
df_1 = df_1.drop(["year","month","day"],axis=1)
df_1
```
