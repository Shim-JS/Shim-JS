```python
sns_temp = df_sns.loc[:,["날짜","내용"]]
sns_temp.head()
```

```python
def count(na):
    #display(df1.Title)
    
    sns_g = sns_temp.groupby("날짜").count()
    sns_g = sns_g.reset_index()
    sns_g = sns_g.rename(columns={"내용" : "SNS"})
    
    k_w = na
    k_c = k_w.split(",")
    sns_k = sns_temp.loc[sns_temp['내용'].str.contains('|'.join(k_c))]
    display(sns_k.head(10))
    sns_k = sns_k.groupby("날짜").count()
    sns_k = sns_k.reset_index()
    sns_k = sns_k.rename(columns={"내용" : "SNS_KeyWord"})
    display(sns_k.head(10))
    
    sns_t = sns_g.copy()
    #sns_t["SNS_KeyWord"] = 0
    sns_t = pd.merge(sns_t,sns_k,how = "left", on = "날짜") 
    sns_t = sns_t.fillna(0)
    
    #df1.Date = pd.to_datetime(df1.Date)
    start = pd.to_datetime("2017-08-17")
    end = pd.to_datetime("2021-08-17")
    sns_t.날짜 = pd.to_datetime(sns_t.날짜)
    
    sns_t = sns_t.loc[sns_t["날짜"].values >= start]
    sns_t = sns_t.loc[sns_t["날짜"].values <= end]
    
    #sns_t["분류"] = df1.Class
    #sns_t["내용"] = df1.Title
    
    name = na 
    sns_t.to_csv("{}_{}.csv".format(name,today),encoding = "cp949",index = False)
    
    return sns_t
```
