```python
df = pd.read_csv("original_data/국민청원_최종_20170817_20210817.csv",encoding = 'cp949')
df
```

```python

for i in df.Class.unique():
    temp = df.loc[df.Class == i]
    temp = temp.sort_values(by = "Support",ascending = False)
    temp = temp.head(3)
    temp.Date = pd.to_datetime(temp.Date)
    temp["기간A시작"] = temp.Date -  dt.timedelta(days=61)
    temp["기간B시작"] = temp.Date - dt.timedelta(days=31)
    temp["기간C시작"] = temp.Date
    temp["기간C종료"] = temp.Date + dt.timedelta(days=31)
    display(temp)
```
