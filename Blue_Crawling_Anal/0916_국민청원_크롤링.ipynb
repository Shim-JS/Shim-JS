```python
####### 패키지 설치 코드 #########
#######'#' 지우고 실행 하면 패키지 설치 #########
#!pip install selenium
#!pip install webdriver_manager

# 패키지 호출
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time
```

```python
today = '0916' 
```

```python
def crawl(start, end):
    # 크롤링한 데이터 저장할 데이터 프레임 생성
    df = pd.DataFrame(columns = ["Class","Title","Date","Support"])
    # 크롤링한 데이터 저장시 필요한 row 변경 x
    row = 0


    # 웹크롤링용 드라이버 설치
    driver = webdriver.Chrome(ChromeDriverManager().install())

    # try, except는 오류 발생시 오류 내용 불러주는 코드 
    try:
        # 페이지별 자동 크롤링 start 시작페이지, end 마지막페이지
        for page in range(start,end):
            # 크롤링할 웹사이트 주소 {} <-여기에 페이지 번호가 삽입
            url = 'https://www1.president.go.kr/petitions/?c=0&only=2&page={}&order=2'.format(page)
            # url 드라이버에 넣고 웹사이트 열기
            driver.get(url)
            # 그 웹 사이트에 특징적인 클래스 이름, ( 크롤링 원하는 클래스 이름 F12 사용해서 웹사이트 분석 후 가지고 오기)
            elem = driver.find_element_by_class_name('b_list.category.b_list2')
            
            # 밑에는 웹사이트 구성별로 다름 (F12 사용해서 웹사이트 분석 후 가지고 오기)
            for num in range(1,8):
                li = '//*[@id="cont_view"]/div[2]/div/div/div[2]/div[2]/div[4]/div/div[2]/div[2]/ul/li[{}]'.format(num)
                # 리스트 형태로 저장했다가
                list_val = list()
                list_val.append(elem.find_elements_by_xpath(li+'/div/div[1]')[0].text)
                list_val.append(elem.find_elements_by_xpath(li+'/div/div[2]')[0].text)
                list_val.append(elem.find_elements_by_xpath(li+'/div/div[3]')[0].text)
                list_val.append(elem.find_elements_by_xpath(li+'/div/div[4]')[0].text)
#                 list_val.append(elem.find_elements_by_xpath(li+'/div/div[5]')[0].text)
                #데이터 프레임에 저장
                df.loc[row] = list_val
                row += 1
            time.sleep(1)

    except Exception as e:
        print(e)
    finally:
        driver.quit()
    return df
```

```python
# 오류 발생으로 3개씩 끊어서 진행
df_final = pd.DataFrame(columns = ["Class","Title","Date","Support"])
for i in range(1,2500,3):
    print(i)
    df_final = pd.concat([df_final, crawl(i,i+3)],axis=0)
```

```python
df_final.to_csv("0817/blue_text_{}_1_2500.csv".format(today),encoding = 'utf-8',index = False)
```
